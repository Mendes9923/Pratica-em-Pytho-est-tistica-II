```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# Parâmetros da distribuição binomial
n = 10  # Número de tentativas
p = 0.5  # Probabilidade de sucesso
```

```python
# Valores possíveis de sucessos
x = np.arange(0, n+1)
```

```python
# Probabilidades da distribuição binomial
binom_pmf = stats.binom.pmf(x, n, p)
```

```python
# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(x, binom_pmf, color='skyblue', edgecolor='black')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.title('Distribuição Binomial (n=10, p=0.5)')
plt.grid(axis='y')
# Adicionando os valores de probabilidade acima de cada barra
for i in range(len(x)):
    plt.text(x[i], binom_pmf[i] + 0.005, f'{binom_pmf[i]:.3f}', ha='center', va='bottom')
    # Salvando o gráfico
plt.savefig('distribuicao_binomial.png')
plt.show()
```
