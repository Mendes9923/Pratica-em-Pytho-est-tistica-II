```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Dados de exemplo para a tabela de contingência
observed = np.array([[30, 10], [10, 50]])  # Frequências observadas
expected = np.array([[20, 20], [20, 40]])  # Frequências esperadas

# Criando um DataFrame para a tabela de contingência
df = pd.DataFrame(observed, columns=['Produto A', 'Produto B'], index=['Gênero X', 'Gênero Y'])

# Calculando a diferença entre observados e esperados
diff = observed - expected

# Criando o gráfico de barras
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Tabela de contingência
sns.heatmap(df, annot=True, fmt='d', cmap='Blues', cbar=False, ax=ax[0])
ax[0].set_title('Tabela de Contingência')
ax[0].set_xlabel('Produto')
ax[0].set_ylabel('Gênero')

# Gráfico de barras para diferença entre observados e esperados
bar_width = 0.35
index = np.arange(len(df.columns))
bar1 = plt.bar(index, observed[0], bar_width, label='Gênero X - Observado')
bar2 = plt.bar(index + bar_width, expected[0], bar_width, label='Gênero X - Esperado', alpha=0.7)
bar3 = plt.bar(index + 2 * bar_width, observed[1], bar_width, label='Gênero Y - Observado')
bar4 = plt.bar(index + 3 * bar_width, expected[1], bar_width, label='Gênero Y - Esperado', alpha=0.7)

ax[1].set_xlabel('Produto')
ax[1].set_ylabel('Frequência')
ax[1].set_title('Diferença entre Observados e Esperados')
ax[1].set_xticks(index + 1.5 * bar_width)
ax[1].set_xticklabels(df.columns)
ax[1].legend()

# Ajustando layout e salvando o gráfico
plt.tight_layout()
plt.savefig('qui_quadrado.png', bbox_inches='tight')

plt.show()
```
