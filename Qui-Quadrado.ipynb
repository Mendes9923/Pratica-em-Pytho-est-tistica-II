```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Dados de exemplo para a tabela de contingncia
observed = np.array([[30, 10], [10, 50]])  # Frequncias observadas
expected = np.array([[20, 20], [20, 40]])  # Frequncias esperadas

# Criando um DataFrame para a tabela de contingncia
df = pd.DataFrame(observed, columns=['Produto A', 'Produto B'], index=['Gnero X', 'Gnero Y'])

# Calculando a diferena entre observados e esperados
diff = observed - expected

# Criando o grfico de barras
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Tabela de contingncia
sns.heatmap(df, annot=True, fmt='d', cmap='Blues', cbar=False, ax=ax[0])
ax[0].set_title('Tabela de Contingncia')
ax[0].set_xlabel('Produto')
ax[0].set_ylabel('Gnero')

# Grfico de barras para diferena entre observados e esperados
bar_width = 0.35
index = np.arange(len(df.columns))
bar1 = plt.bar(index, observed[0], bar_width, label='Gnero X - Observado')
bar2 = plt.bar(index + bar_width, expected[0], bar_width, label='Gnero X - Esperado', alpha=0.7)
bar3 = plt.bar(index + 2 * bar_width, observed[1], bar_width, label='Gnero Y - Observado')
bar4 = plt.bar(index + 3 * bar_width, expected[1], bar_width, label='Gnero Y - Esperado', alpha=0.7)

ax[1].set_xlabel('Produto')
ax[1].set_ylabel('Frequncia')
ax[1].set_title('Diferena entre Observados e Esperados')
ax[1].set_xticks(index + 1.5 * bar_width)
ax[1].set_xticklabels(df.columns)
ax[1].legend()

# Ajustando layout e salvando o grfico
plt.tight_layout()
plt.savefig('qui_quadrado.png', bbox_inches='tight')

plt.show()
```
