```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
 #Parâmetro da distribuição Poisson
lambda_ = 3  # Taxa média de ocorrência

# Valores possíveis de eventos
x = np.arange(0, 11)

# Probabilidades da distribuição Poisson
poisson_pmf = stats.poisson.pmf(x, lambda_)


```

```python
# Criando o gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(x, poisson_pmf, color='skyblue', edgecolor='black')
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidade')
plt.title('Distribuição Poisson (λ=3)')
plt.grid(axis='y')
# Adicionando os valores de probabilidade acima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, f'{yval:.3f}', ha='center', va='bottom')

# Adicionando a legenda explicando os eixos
plt.figtext(0.15, -0.05, 'O eixo X representa o número de eventos em um intervalo. O eixo Y representa a probabilidade de cada número de eventos.', wrap=True, horizontalalignment='left', fontsize=10)

# Salvando o gráfico
plt.savefig('distribuicao_poisson_com_legenda.png', bbox_inches='tight')

plt.show()
```
