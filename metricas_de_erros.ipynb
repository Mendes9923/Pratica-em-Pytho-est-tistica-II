```python
import matplotlib.pyplot as plt
```

```python
# Dados para o infográfico
metrics = ['MAE', 'MSE', 'RMSE', 'MAPE']
definitions = [
    'Erro Médio Absoluto: Média das diferenças absolutas entre previsões e valores reais.',
    'Erro Quadrático Médio: Média das diferenças quadráticas, penalizando erros maiores.',
    'Raiz do Erro Quadrático Médio: Raiz quadrada do MSE, na mesma escala dos dados originais.',
    'Erro Percentual Absoluto Médio: Média das diferenças absolutas em termos percentuais.'
]
formulas = [
    r'$\frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|$',
    r'$\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$',
    r'$\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$',
    r'$\frac{1}{n}\sum_{i=1}^{n}\left|\frac{y_i - \hat{y}_i}{y_i}\right| \times 100$'
]

# Criando o infográfico
fig, ax = plt.subplots(figsize=(12, 8))

# Removendo eixos
ax.axis('off')

# Adicionando título e subtítulo
plt.text(0.5, 0.95, 'Métricas de Erros', fontsize=24, ha='center', va='center', fontweight='bold')
plt.text(0.5, 0.9, 'Avaliando a Precisão dos Modelos', fontsize=18, ha='center', va='center')

# Adicionando as métricas
for i, (metric, definition, formula) in enumerate(zip(metrics, definitions, formulas)):
    plt.text(0.1, 0.75 - i*0.2, metric, fontsize=20, ha='left', va='center', fontweight='bold', color='blue')
    plt.text(0.1, 0.72 - i*0.2, definition, fontsize=14, ha='left', va='center')
    plt.text(0.1, 0.69 - i*0.2, formula, fontsize=14, ha='left', va='center', color='darkred')

# Ajustando layout e salvando o infográfico
plt.tight_layout()
plt.savefig('metricas_de_erros.png', bbox_inches='tight')

plt.show()
```
